```python
# prompt: plot code ti covert category data ti numirc

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your code from the previous response is already executed and 'df' is loaded

# ... (your existing code)

# Function to convert categorical data to numerical using label encoding
def encode_categorical_features(df):
    """
    Encodes categorical features in a pandas DataFrame using label encoding.

    Args:
        df: The input DataFrame.

    Returns:
        A new DataFrame with categorical features converted to numerical.
    """
    encoded_df = df.copy()  # Create a copy to avoid modifying the original DataFrame
    for col in encoded_df.select_dtypes(include=['object']):
        encoded_df[col] = encoded_df[col].astype('category').cat.codes
    return encoded_df


# Apply label encoding to the categorical features in df
df_encoded = encode_categorical_features(df)


# Example visualization (you can choose any pair of columns):
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_encoded, x="Sleep Disorder", y="BMI Category", hue='Occupation')
plt.title("Relationship Between Sleep Disorder and BMI Category")
plt.xlabel("Sleep Disorder (Encoded)")
plt.ylabel("BMI Category (Encoded)")
plt.show()

# Show the first few rows of the encoded DataFrame
print(df_encoded.head())
```

```python
# prompt: plot code to find correlation betwean data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df_encoded is already created from previous code

# Calculate the correlation matrix
correlation_matrix = df_encoded.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Encoded Features')
plt.show()

# Find the correlation between two specific columns (example: 'Sleep Disorder' and 'BMI Category')
correlation_value = correlation_matrix.loc['Sleep Disorder', 'BMI Category']
print(f"Correlation between 'Sleep Disorder' and 'BMI Category': {correlation_value}")
```

```python
# prompt: plot chart for max ,min and average age

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df_encoded is already created from the previous code

# Calculate max, min, and average age
max_age = df_encoded['Age'].max()
min_age = df_encoded['Age'].min()
avg_age = df_encoded['Age'].mean()

# Create a bar plot
age_stats = pd.DataFrame({
    'Statistic': ['Max Age', 'Min Age', 'Average Age'],
    'Value': [max_age, min_age, avg_age]
})

plt.figure(figsize=(8, 6))
sns.barplot(x='Statistic', y='Value', data=age_stats)
plt.title('Max, Min, and Average Age')
plt.ylabel('Age')
plt.show()
```

```python
# prompt: plot chart betwean group by  age and Gender

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataframe
df = pd.read_csv('/content/sleep_health_lifestyle_dataset.csv', sep=',')

# Group by 'Age' and 'Gender' and count the occurrences
age_gender_counts = df.groupby(['Age', 'Gender']).size().reset_index(name='Count')

# Create the plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Age', y='Count', hue='Gender', data=age_gender_counts)
plt.title('Distribution of Gender across different Age Groups')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout() #adjust layout to prevent labels from overlapping
plt.show()
```
